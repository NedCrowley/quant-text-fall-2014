### Topic Models in Python

There are four main ways to estimate topic models in Python:

| Name   | Methodology                                        | Installation                |
|--------|----------------------------------------------------|-----------------------------|
| [gensim](http://radimrehurek.com/gensim/) | LDA using  online variational Bayes (VB) algorithm | pip install gensim                         |
| [lda](https://pypi.python.org/pypi/lda)    | LDA using collapsed Gibbs sampling                 | pip install lda                        |
| [Mallet](http://mallet.cs.umass.edu)  | LDA suing collapsed Gibbs sampling                 | [Java download.](http://mallet.cs.umass.edu/download.php) <br>Requires JDK and  [gensim](http://radimrehurek.com/gensim/models/ldamallet.html) wrapper. |
| [NMF](http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.NMF.html)    | Non-Negative matrix factorization                  | part of [scikit-learn](http://scikit-learn.org/stable/)        |

```python
# Note that this will only work if you have stored the data in the current directory. 
# You can get the data from  
# https://github.com/nealcaren/quant-text-fall-2014/tree/master/times_data
# Remember that Pandas can read a file or a URL.

import pandas as pd



df = pd.read_json('times_abortion.json')

print 'Article Count:', len(df)
print 'Word Count:', df.word_count.sum()
df.head()
```

```python
# Trim the dataset to just look at a couple of newsdesks

def good_desk(dsk):
    if dsk in ['National Desk', 'Editorial Desk',  u'Metropolitan Desk', 'Foreign Desk']:
        return True
    return False

df['viable'] = df['dsk'].apply(good_desk)

df = df.loc[df['viable'] == True]
```

```python
df.groupby('dsk').count()
```

```python
print len(df)
df.head()
```

```python
from datetime import datetime

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
tf_vectorizer = CountVectorizer(max_df=0.7, 
                             min_df=.01,
                             stop_words='english')


tf_vectorizer.fit(df['text'])
corpus_tf = tf_vectorizer.transform(df['text'])
```

```python
categories = list(set(df.dsk.unique()))

number_of_topics = len(categories)  * 4

print 'Planning to extract %s topics' % number_of_topics
```

```python
from sklearn.decomposition import NMF


start_time = datetime.now()

nmf = NMF(n_components = number_of_topics, 
          max_iter = 5000).fit(corpus_tf)

elapsed_time = (datetime.now() - start_time).seconds
print elapsed_time, 'seconds'
```

```python
feature_names = tf_vectorizer.get_feature_names()
topic_labels = []

for topic_idx, topic in enumerate(nmf.components_):
    print "Topic %s: %s" % (topic_idx,  ' '.join([feature_names[i] for i in topic.argsort()[:-10 - 1:-1]]))
    topic_labels.append(" ".join([feature_names[i]
                    for i in topic.argsort()[:-3 - 1:-1]]))
```

```python
y_hat = nmf.transform(corpus_tf)
y_norm = [y/ y.sum() for y in y_hat] #Normalize

df_y_hat = pd.DataFrame(y_norm, 
                        columns=topic_labels,  # Use our informative topic names
                        index=df.index.values) # Makes merge easier

df_y_hat = df_y_hat.fillna(0)
df_topics = df.join(df_y_hat) # This is a new data set, so you can remerge
```

```python
df_topics[['dsk'] + topic_labels].groupby('dsk').mean()
```

```python
model_stats = {}
model_stats['NMF'] = {'time' : elapsed_time ,
                      'labels' : topic_labels ,
                      'dataframe' : df_topics }
```

```python
from gensim import corpora, models, similarities, matutils 
```

```python
id2word = dict((v, k) for k, v in tf_vectorizer.vocabulary_.iteritems())
corpus_vect_gensim = matutils.Sparse2Corpus(corpus_tf, documents_columns=False)
```

```python
start_time = datetime.now()

lda = models.LdaModel(corpus_vect_gensim, 
                      id2word = id2word, 
                      num_topics = number_of_topics,
                      passes=10)
elapsed_time = (datetime.now() - start_time).seconds
print elapsed_time, 'seconds'
```

```python
topic_labels = []


for i in range(0, number_of_topics):
    temp = lda.show_topic(i, 10)
    terms = []
    for term in temp:
        terms.append(term[1])
    print  "topic %s: %s" % (i," ".join(terms) )
    topic_labels.append(" ".join(terms[:3]))
```

```python
y_hat = []
for item in corpus_vect_gensim:
    y_dict = dict(lda[item])
    y_list = [y_dict.get(i,0) for i in range(0,number_of_topics)]
    y_hat.append(y_list)

df_y_hat = pd.DataFrame(y_hat, 
                        columns=topic_labels,  # Use our informative topic names
                        index=df.index.values) # Makes merge easier

df_topics = df.join(df_y_hat) # This is a new data set, so you can remerge
```

```python
df_topics[['dsk'] + topic_labels].groupby('dsk').mean()
```

```python
model_stats['Gensim'] = {'time' : elapsed_time ,
                         'labels' : topic_labels ,
                         'dataframe' : df_topics }
```

```python
import lda
import numpy as np
```

```python
vocab = [w for w in sorted(tf_vectorizer.vocabulary_)]


model = lda.LDA(n_topics= number_of_topics,
                n_iter=500)

```

```python
start_time = datetime.now()

model.fit(corpus_tf.toarray())

elapsed_time = (datetime.now() - start_time).seconds
print elapsed_time, 'seconds'
```

```python
topic_word = model.topic_word_
topic_labels = []

n_top_words = 10
for i, topic_dist in enumerate(topic_word):
    
    topic_words = np.array(vocab)[np.argsort(topic_dist)][:-n_top_words:-1]
    print('Topic {}: {}'.format(i, ' '.join(topic_words)))
    topic_labels.append(' '.join(topic_words[:3]))
```

```python
df_y_hat = pd.DataFrame(model.doc_topic_, 
                        columns=topic_labels,  # Use our informative topic names
                        index=df.index.values) # Makes merge easier

df_topics = df.join(df_y_hat) # This is a new data set, so you can remerge
```

```python
df_topics[['dsk'] + topic_labels].groupby('dsk').mean()
```

```python
model_stats['LDA'] = {'time' : elapsed_time ,
                      'labels' : topic_labels ,
                      'dataframe' : df_topics }
```

```python
from string import punctuation


vocab = tf_vectorizer.vocabulary_
texts = [[word.strip(punctuation) for word in document.lower().split() if word in vocab ] for document in df['text']]

dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
mallet_path = '/Users/nealcaren/Downloads/mallet-2.0.7/bin/mallet'

start_time = datetime.now()

model = models.LdaMallet(mallet_path, corpus, num_topics=number_of_topics, id2word=dictionary)

elapsed_time = (datetime.now() - start_time).seconds
print elapsed_time, 'seconds'
```

```python
topic_labels = []

for i in range(0, number_of_topics):
    temp = model.show_topic(i, 10)
    terms = []
    for term in temp:
        terms.append(term[1])
    print "topic %s: %s" % (i," ".join(terms))
    topic_labels.append(" ".join(terms[:3]))

```

```python
y_hat = []
predictions = model[corpus]
for item in predictions:
    y_dict = dict(item)
    y_list = [y_dict.get(i,0) for i in range(0,number_of_topics)]
    y_hat.append(y_list)

df_y_hat = pd.DataFrame(y_hat, 
                        columns=topic_labels,  # Use our informative topic names
                        index=df.index.values) # Makes merge easier

df_topics = df.join(df_y_hat) # This is a new data set, so you can remerge
```

```python
df_topics[['dsk'] + topic_labels].groupby('dsk').mean()
```

```python
model_stats['MALLET'] = {'time' : elapsed_time ,
                      'labels' : topic_labels ,
                      'dataframe' : df_topics }

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report 

%matplotlib inline
import matplotlib.pyplot as plt


lr = LogisticRegression()
```

```python
def cmplot(y_true, y_pred, labels):
    print classification_report(y_true, y_pred, labels)
    cm = confusion_matrix(y_true, y_pred, labels)
    correct = accuracy_score(y_true,y_pred )*100
    fig = plt.figure()
    ax = fig.add_subplot(111)
    cax = ax.matshow(cm)
    plt.title('Confusion matrix of the classifier\n %.1f%% correctly classified' %  correct)
    fig.colorbar(cax)
    ax.set_xticklabels([''] )
    ax.set_yticklabels([''] + categories)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

```

```python
for method in sorted(model_stats):
    print '%s took %s seconds.' % (method, model_stats[method]['time'])
    predictors = model_stats[method]['labels'][:-1]
    df = model_stats[method]['dataframe']
    
    lr.fit(df[predictors], df['dsk'] )
    
    prediction = lr.predict(df[predictors])
    cmplot(df['dsk'], prediction, categories)
    print ''
    
    
```
