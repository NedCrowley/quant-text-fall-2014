```python
%matplotlib inline

import pandas as pd

import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
```

```python
# Note that this will only work if you have stored the data in the current directory. 
# You can get the data from  
# https://github.com/nealcaren/quant-text-fall-2014/tree/master/times_data
# Remember that Pandas can read a file or a URL.

df = pd.read_json('times_abortion.json')

print 'Article Count:', len(df)
print 'Word Count:', df.word_count.sum()
df.head()
```

```python
dfn = pd.read_json('times_national_abortion.json')

print 'Article Count:', len(dfn)
print 'Word Count:', dfn.word_count.sum()
df.head()
```

```python
# LDA uses frequencies, but I find that NMF likes Tf-Idf better.

vectorizer = TfidfVectorizer(max_df=0.7, 
                             min_df=5,
                             stop_words='english')
```

```python
# This may not be quick. The whole notebook takes my relatively modern MacBook Air 5 minutes to run.

vectorizer.fit(df['text'])
tfidf = vectorizer.transform(df['text'])
```

```python
# This will not be quick. Note that more topics = more time.

number_of_topics = 20

nmf = NMF(n_components = number_of_topics, max_iter = 5000).fit(tfidf)
```

```python
# Print out most informative features 
# This also stores the topics in a list topic_labels



feature_names = vectorizer.get_feature_names()
topic_labels = []

for topic_idx, topic in enumerate(nmf.components_):
    print("Topic #%d:" % topic_idx)
    print(" ".join([feature_names[i]
                    for i in topic.argsort()[:-15 - 1:-1]]))
    topic_labels.append(" ".join([feature_names[i]
                    for i in topic.argsort()[:-3 - 1:-1]]))
    print ''
```

```python
# Get some predicted values. Takes a tfidf.

y_hat = nmf.transform(tfidf)
```

```python
# get the most likely topic and add it as a column

topics = []
for row in y_hat:
    topics.append(topic_labels[np.argsort(row)[0]])
df['top_topic'] = topics
```

```python
# get the full suite of predictions as values
# This is normalized so that the topic values for a given article sum to 1

y_norm = [y/ y.sum() for y in y_hat] #Normalize

df_y_hat = pd.DataFrame(y_norm, 
                        columns=topic_labels,  # Use our informative topic names
                        index=df.index.values) # Makes merge easier

df_topics = df.join(df_y_hat) # This is a new data set, so you can remerge
```

```python
#an example of the stats could do

for topic in topic_labels:
    print df_topics[topic].mean(), topic
```
