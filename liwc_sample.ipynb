```python
import requests
import json
```

```python
try:
    import liwc
except Exception, e:
    import codecs
    url = 'http://goo.gl/i6I54j'
    with codecs.open('liwc.py', 'w', 'utf-8') as outfile:
        text = requests.get(url).text
        outfile.write(text)
    import liwc
```

```python
# try and load up the emails. If that doesn't work, get them from the internet.

try:
    emails =json.load(open('emails.json', 'r'))
except exception, e:
    url = 'https://raw.githubusercontent.com/nealcaren/quant-text-fall-2014/master/emails.json'
    emails = requests.get(url).json()
    #save them for next time
    json.dump(open('emails.json','w'))
```

```python

# Loop through each of the emails. Grab some useful features for each, 
# including the results of a liwc analysis - liwc.liwcify()

processed_emails = []

for email in emails:
    text = email['copy'].replace("If this isn't the best email address at which to reach you, update your contact information. Our email list is the best way we have of staying in regular contact with supporters like you across the country and letting you know about the work President Obama and other Democrats are doing. Click here to unsubscribe from our supporter list. This organization is powered by you, and we'd love to hear your ideas. Send us any comments, criticisms, or feedback here. Thanks for supporting President Obama and other Democrats.",'')
    processed = liwc.liwcify(text, normalize=True)
    processed['year'] = int(email['date'].split()[2])
    processed['sender'] = email['sender'].split('@')[-1].split('>')[0].replace('email.','')
    processed_emails.append(processed)
```

```python
#Save the results to a CSV, using Pandas because its quick and easy
import pandas as pd

df = pd.DataFrame(processed_emails).fillna(0)
df.to_csv('email_liwcs.csv', index=False)
```
